```python
%sql
-- Cambiá la celda del notebook a %sql
SELECT 
    tipo_operacion,
    COUNT(id_transaccion) AS cantidad_operaciones,
    ROUND(SUM(monto_total), 2) AS volumen_financiero_total
FROM platform_dev.gold_byma.fact_operaciones
GROUP BY tipo_operacion;
```

```python
%sql
SELECT 
    i.simbolo,
    i.descripcion,
    ROUND(SUM(f.monto_total), 2) AS volumen_total_pesos
FROM platform_dev.gold_byma.fact_operaciones f --
LEFT JOIN platform_dev.gold_byma.dim_instrumento i ON f.instrumento_sk = i.instrumento_sk --
GROUP BY i.simbolo, i.descripcion
ORDER BY volumen_total_pesos DESC
limit 10;
```

```python
%sql
SELECT 
    f.fecha,
    ROUND(SUM(fact.monto_total), 2) AS volumen_diario
FROM platform_dev.gold_byma.fact_operaciones fact
LEFT JOIN platform_dev.gold_byma.dim_fecha f ON fact.fecha_sk = f.fecha_sk
GROUP BY f.fecha
ORDER BY f.fecha ASC;
```

```python
%sql
SELECT 
    tipo_operacion,
    COUNT(CASE WHEN flag_compro_caro = true THEN 1 END) AS alertas_compro_caro,
    COUNT(CASE WHEN flag_vendio_barato = true THEN 1 END) AS alertas_vendio_barato,
    ROUND(AVG(ABS(desvio_pct)), 2) AS desvio_promedio_mercado_pct
FROM platform_dev.gold_byma.fact_operaciones
GROUP BY tipo_operacion;
```

```python
%sql
SELECT 
    tipo_operacion,
    COUNT(CASE WHEN flag_compro_caro = true OR flag_vendio_barato = true THEN 1 END) AS operaciones_con_alerta,
    COUNT(CASE WHEN flag_compro_caro = false AND flag_vendio_barato = false THEN 1 END) AS operaciones_normales
FROM platform_dev.gold_byma.fact_operaciones --
GROUP BY tipo_operacion;
```
